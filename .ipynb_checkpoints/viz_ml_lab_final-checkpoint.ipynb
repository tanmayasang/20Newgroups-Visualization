# Getting Data from Sklearn 

```python
# Fetching the data
from sklearn.datasets import fetch_20newsgroups
import numpy as np
newsgroups = fetch_20newsgroups(subset='test')

# getting the topic ids
topic_idx = np.array(newsgroups.target, dtype=int)

# getting the unique topic names
topic_names = np.array(newsgroups.target_names)

# getting the list of documents
documents = list(newsgroups.data)

# getting the list of topics (in the same order as documents)
topics = list(topic_names[topic_idx])
```

# Preprocessing the data

```python
#function to preprocess the document
def preprocess_data(documents):
    docs_list = []
    
    for text in documents:
        #convert text to lower case
        text = text.lower()
        #tokenize
        text = word_tokenize(text)  
        #remove stopwords and single letters
        text = [word for word in text if word not in stop_words and len(word) > 1]
        #Append text list to text_list
        docs_list.append(text)
    
    return docs_list
```

```python
#preprocess the documents
documents_new = preprocess_data(documents)
```

```python
#build a dataframe 
df_list = [list([topics[i], documents_new[i]]) for i in range(len(documents_new))]
df = pd.DataFrame(df_list, columns=['topic', 'words'])
df
```

```python
#function to return the top k words
from collections import Counter
def top_k_words(dff, k):
    word_series = dff['words']
    li=[]
    for element in (word_series):
        li.extend(element)
    most_common=Counter(li).most_common(k)
    top_k = []
    
    for element in most_common:
        t = element
        # create a format with which D3.js will be able to make a bar chart 
        top_k.append({"topic":t[0],"frequency":t[1]})
    return top_k 
```

# Sending data 

```python
import random
#global variable to keep track of selected option--> to be used later in get_Exported_documents function 
doclist_topic="dummytopic"
#function to return data for barplot
def target_func(comm, open_msg):
    
    # comm is the kernel Comm instance

    # Register handler for later messages
    @comm.on_msg
    def _recv(msg):
        # Use msg['content']['data'] for the data in the message
        selected_topic = msg['content']['data']['optionselect']
        selected_topic = str(selected_topic) 
        fil_data=df[df.topic.isin([selected_topic])]
        k=5
        data_words= top_k_words(fil_data, k)
        comm.send({'array': data_words})
        
def target_func2(comm, open_msg):
    # comm is the kernel Comm instance

    # Register handler for later messages
    @comm.on_msg
    def _recv(msg):
        # Use msg['content']['data'] for the data in the message
        global doclist_topic
        top = msg['content']['data']['topp']
        top = str(top) 
        #update the global variable if button gets clicked
        doclist_topic = top
        
get_ipython().kernel.comm_manager.register_target('my_comm_target', target_func)
get_ipython().kernel.comm_manager.register_target('my_comm_target2', target_func2)

    
```

```python
#call this function when you click the Export Documents button
def get_exported_documents():
    #dataframe of selected topic documents
    documents_dataframe = df[df.topic.isin([doclist_topic])]
    #create a new column which convers each document to a string 
    documents_dataframe['liststring']= [' '.join(map(str, l)) for l in documents_dataframe['words']]    
    #convert to list datatype since we have to return a list of strings
    return (documents_dataframe['liststring']).values.tolist()
```
